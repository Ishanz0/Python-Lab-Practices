```python
def win():
  computer = random.choice(['rock','paper','scissor'])
  player = random.choice(['rock','paper','scissor'])
  while computer==player:
    computer=random.choice(['rock','paper','scissor'])
  print("Player choiced ", player)
  print("Computer Choiced ", computer)
  if(player == "paper"):
    if(computer == "rock"):
      print("Player wins")
    else:
      print("Computer wins")
  elif(player == "scissor"):
    if(computer == "paper"):
      print("Player wins")
    else:
      print("Computer wins")
  elif(player == "rock"):
    if(computer == "scissor"):
      print("Player wins")
    else:
      print("Computer wins")
```

```python
def a():
  p1=0
  p2=0
  snakes={97:78,95:56,88:24,68:18,48:26,36:6,32:10}
  ladders={1:38,4:14,8:30,21:42,28:76,50:67,71:92,80:99}
                         
  while p1<100 and p2<100:
    choice=int(input("Enter 1 to throw dice\n"))
    if choice==1:
      p1d=random.randint(1,6)
      if 100-p1<p1d:
        print("Errr \n")

        p2d=random.randint(1,6)
        if 100-p2<p2d:
          continue
        p2+=p2d
        if p2 in snakes.keys():
          p2=snakes[p2]
        if p2 in ladders.keys():
          p2=ladders[p2]
        print("Computer position ",p2)
        continue

      print("dice shows ",p1d)
      p1+=p1d
      if p1 in snakes.keys():
        print("Snake bites!, Slips from ",p1," to ",snakes[p1])
        p1=snakes[p1]
      if p1 in ladders.keys():
        print("Its ladder, Climb up from ",p1," to ",ladders[p1])
        p1=ladders[p1]
      print("Your position ",p1)

      p2d=random.randint(1,6)
      if 100-p2<p2d:
        continue
      p2+=p2d
      if p2 in snakes.keys():
        p2=snakes[p2]
      if p2 in ladders.keys():
        p2=ladders[p2]
      print("Computer position ",p2)
    
  if p1==100:
    print("User Won")
    return True
  else:
    print("User Lost")
    return False
```

```python
def flames():
  a=input("enter name ").lower()
  b=input("enter partners's name ").lower()
  c = a+b
  for d in c:
    if c.count(d) != 1:
      c = c.replace(d,"")

  n = len(c) % 6
  r =""

  if n == 1:
    r = "Friends"
  elif n == 2:
    r = "Love"
  elif n == 3:
    r = "Attraction"
  elif n == 4:
    r = "Marraige"
  elif n == 5:
    r = "Enemy"
  elif n == 6:
    r ="siblings"
  print(r)
```

```python
def m():
  movie=["extraction","kgf","pushpa"]
  name=random.choice(movie)
  name=name.lower()
  name.replace(" ","")
  print(name)
  n=len(name)
  ans=[]
  for i in range(n):
    ans.append("*")
  print(ans)

  counter=0
  chance=3
  while counter<n and chance>0:
    ch=input("Guess Character :")
    flag=False
    for j in range(n):
      if ch==name[j]:
        ans[j]=ch
        counter=counter+1
        print(ans)
        flag=True
    if(flag==False):
      chance=chance-1
      print("You guessed Wrong, remaining chances :",chance)
  if chance<=0:
    print("YOU LOST NOOB")
    return False
  else:
    print("YOU WON")
    return True
```
